```python
def get_lowest(uam_to_check=""):
    rawData         = np.genfromtxt(uam_to_check)[:,0:7]
    phiData         = rawData[:,0]
    thetaData       = rawData[:,1]
    #energy          = rawData[:,2] #kbt
    energy          = rawData[:,6] #kJ/mol
    SD             = 2.49* rawData[:,3] #kJ/mol

    sinTheta        = np.sin((2.5+thetaData) * np.pi / 180.0)

    boltzmanAverage = np.sum(energy * sinTheta * np.exp(-1.0 *  energy)) / np.sum(sinTheta * np.exp(-1.0 * energy))
    simpleAverage   = np.sum(sinTheta * energy)/np.sum(sinTheta) # corrected so that it works properly for isotropic proteins and single AAs with E(phi,theta) = E, else we get Esimple =  2E/Pi 
    index           = np.argmin(energy)
    return energy[index], phiData[index],  thetaData[index], SD[index], boltzmanAverage
```

```python
def heatmapAnalyse(moda ='attraction', rank=0.3):
    
    if (moda == 'attraction'):
        file_tail="_attraction_poses.dat"
    if (moda == 'repulsion'):
        file_tail="_repulsion_poses.dat"      
    outfile=resFolder+uamIn.rsplit(".",1)[0].rsplit("/")[-1]+file_tail
       
    df = pd.DataFrame({"file":outfile, 
                       "phi":rawData[:,0],
                       "theta": rawData[:,1],
                       "sd": 2.459*rawData[:,3],
                       "ssd": rawData[:,4], 
                       "ene":rawData[:,6]})#kJ/mol
    
    triplex, ax = plt.subplots(1, 3, figsize=(5, 5))#, sharex=True, sharey=True)
    #ax = axes.ravel()

    
    #Plot original
    
    original           = np.zeros((rows, cols))
    for phi, theta ,energy0 ,_ ,_, _ ,_, _, _, _, _   in rawData:
        row             = int(theta / delta)
        col             = int(phi / delta)
        original[row][col] =  2.459*energy0
        
    #original = sp.ndimage.filters.gaussian_filter(original, sigma = [1.0, 1.0], mode = ['constant', 'wrap'])
    title_0='heatmap -->'
    ax[0].set_title(title_0)
    ax[0].imshow(original, cmap = cmap)

    #Get transform    
    
    filter_size=0.5
    filter_dist=2
    rank=rank #sampling 10%
    dataIn=rawData

    if (moda == 'repulsion'):
        dataIn[:,2] = 1.0* 2.459*dataIn[:,2] # keep original
        title_1=' *(1.0)- -> transformed'
        title_2=' -> peaks location (repulsion)'
    else:
            dataIn[:,2] = -1.0* 2.459*dataIn[:,2] # create negative
            title_1='  *(-1.0) --> transformed'
            title_2=' -> peaks location (attraction)'
            
         
    landscape_max=max(dataIn[:,2])
    landscape_min=min(dataIn[:,2])
    threshold = landscape_max - ((landscape_max-landscape_min)*rank)

    # Fill out with values
    transform           = np.zeros((rows, cols))
    for phi, theta ,energy ,_ ,_, _ ,_, _, _, _, _   in dataIn:
        row             = int(theta / delta)
        col             = int(phi / delta)
        transform[row][col] = energy
            
    # Smooth and find local minima (attractive orientations) and maxima (repulsive orientations) 
  
    #transformSmooth = ndi.filters.gaussian_filter(transform, sigma = [1.0, 1.0], mode = ['constant', 'wrap'])
    transformMax = ndi.maximum_filter(transform, size=filter_size)   
    coor = peak_local_max(transformMax, min_distance=filter_dist, threshold_abs=threshold)    
    levels =   np.arange(0,10,0.1)
    print(coor)
    
    #Plot transformed heatmap

    ax[1].imshow(transform, cmap = cmap)
    ax[1].set_title(title_1)
    
    #Plot transformed heatmap    + peaks location
    
    ax[2].imshow(transformMax, cmap=cmap)
    ax[2].autoscale(False)
    ax[2].plot(coor[:, 1], coor[:, 0], "blue", marker='o')
    ax[2].set_title(title_2)
    
    triplex.tight_layout()    
    #ax[2].triplex.axes
    
    #Convert locations to angles in degrees
    loc = coor*delta
    dfloc=pd.DataFrame({"phi":loc[:,1], "theta": loc[:,0]})

    
    #Find line with absolute minimum (or maximum)
    if (moda == 'attraction'):
        fst_line =df.iloc[np.where(df.ene.values==min(df["ene"]))]
    else:
        fst_line =df.iloc[np.where(df.ene.values==max(df["ene"]))]
    
    res_out = []
    res_out.append(fst_line)

    for index, row in dfloc.iterrows():
        loc_phi   = row['phi']
        loc_theta = row['theta']
        selection=df.iloc[np.where((df.phi.values==loc_phi) & (df.theta.values==loc_theta))]
        res_out.append(selection)

    res_out= pd.concat(res_out)

    if (moda == 'attraction'):
        res_out.sort_values(by=['ene'], ascending=True )
    if (moda == 'repulsion'):
        res_out.sort_values(by=['ene'], ascending=False)

    res_out.to_csv(outfile, index=False)
    return res_out
```

```python
# adopted Ian code for rotation
def rotatePDB(coords,phi,theta):
    rxx = np.cos(theta)*np.cos(phi)
    rxy = -1 * np.cos(theta)*np.sin(phi)
    rxz = np.sin(theta)
    ryx = np.sin(phi)
    ryy = np.cos(phi)
    ryz = 0
    rzx = -1*np.sin(theta)*np.cos(phi)
    rzy = np.sin(theta)*np.sin(phi)
    rzz = np.cos(theta)
    finalCoords = np.zeros_like(coords)
    
    finalCoords[:,0] = coords[:,0] * rxx + coords[:,1] * rxy + coords[:,2]*rxz 
    finalCoords[:,1] = coords[:,0] * ryx + coords[:,1] * ryy + coords[:,2]*ryz 
    finalCoords[:,2] = coords[:,0] * rzx + coords[:,1] * rzy + coords[:,2]*rzz 
    return finalCoords
```

```python
def tUACOM():
    def wrapped(ts):
        ts.positions += np.array(  [ -np.mean(ts.positions[:,0]) ,- np.mean(ts.positions[:,1]), -np.mean(ts.positions[:,2])  ]  )
        return ts
    return wrapped
       
def tUARotate(phi,theta):
    def wrapped(ts):
        ts.positions = rotatePDB(ts.positions,phi,theta)
        return ts
    return wrapped

def tUAZShift(dz):
    def wrapped(ts):
        ts.positions += np.array( [0.0,0.0,dz - np.amin(ts.positions[:,2])])
        return ts
    return wrapped
```
